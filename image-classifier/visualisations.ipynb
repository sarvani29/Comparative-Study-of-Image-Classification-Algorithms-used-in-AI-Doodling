```python
d = { 
    0:[],
    1:[],
    2:[],
    3:[]
}
with open("model_logs.txt", "r") as f:
    for i, line in enumerate(f):
        split_line = line.split()
        if(i==0):
            header = split_line
        elif(i>=2):
            for j, val in enumerate(split_line):
                d[j].append(float(val))
        
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(d)
```

```python
df.columns = header
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
# plotting line 1 points
plt.plot(list(df.index), df["train_loss"], label = "Train loss")
# plotting the line 2 points 
plt.plot(list(df.index), df["valid_loss"], label = "Test loss")
plt.xlabel('Iterations')
# Set the y axis label of the current axis.
plt.ylabel('Loss')
# Set a title of the current axes.
plt.title('Train and Test Loss')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```

```python
# plotting line 1 points
plt.plot(list(df.index), df["train_accuracy"], label = "Train accuracy")
# plotting the line 2 points 
plt.plot(list(df.index), df["valid_accuracy"], label = "Test accuracy")
plt.xlabel('Iterations')
# Set the y axis label of the current axis.
plt.ylabel('Accuracy')
# Set a title of the current axes.
plt.title('Train and Test Accuracy')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```
